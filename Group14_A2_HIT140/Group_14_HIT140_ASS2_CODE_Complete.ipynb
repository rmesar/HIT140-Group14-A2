```python
# Set dark background
sns.set(style="darkgrid")
plt.style.use("dark_background")

# Function to plot sorted histogram
def plot_sorted_histogram(data, ax, title, color):
    # Calculate the frequency distribution
    freq_dist = data.value_counts().sort_index()

    # Convert to DataFrame for easier plotting
    freq_dist_df = pd.DataFrame({'Value': freq_dist.index, 'Frequency': freq_dist.values})

    # Plot with annotations
    sns.barplot(x='Value', y='Frequency', data=freq_dist_df, ax=ax, color=color)

    # Add annotations for each bar
    for p in ax.patches:
        ax.annotate(f'{p.get_height():.0f}', (p.get_x() + p.get_width() / 2., p.get_height()), 
                    ha='center', va='center', xytext=(0, 5), textcoords='offset points')

    # Set title and labels
    ax.set_title(title, fontsize=14, fontweight='bold')
    ax.set_xlabel('Hours', fontsize=12)
    ax.set_ylabel('Count', fontsize=12)
    
    # Set ticks and add 'Over' for values beyond 7
    ax.set_xticks(range(9))  # Set xticks from 0 to 8 (for Over)
    ax.set_xticklabels([str(i) for i in range(8)] + ['Over'])  # Keep '7' and add 'Over'

# Plotting
fig, axs = plt.subplots(4, 2, figsize=(18, 16))

# Computer use
plot_sorted_histogram(dataset2['C_wk'], axs[0, 0], 'Computer Use (Weekdays)', 'skyblue')
plot_sorted_histogram(dataset2['C_we'], axs[0, 1], 'Computer Use (Weekends)', 'skyblue')

# Gaming
plot_sorted_histogram(dataset2['G_wk'], axs[1, 0], 'Gaming (Weekdays)', 'orange')
plot_sorted_histogram(dataset2['G_we'], axs[1, 1], 'Gaming (Weekends)', 'orange')

# Smartphone use
plot_sorted_histogram(dataset2['S_wk'], axs[2, 0], 'Smartphone Use (Weekdays)', 'green')
plot_sorted_histogram(dataset2['S_we'], axs[2, 1], 'Smartphone Use (Weekends)', 'green')

# TV Watching
plot_sorted_histogram(dataset2['T_wk'], axs[3, 0], 'TV Watching (Weekdays)', 'purple')
plot_sorted_histogram(dataset2['T_we'], axs[3, 1], 'TV Watching (Weekends)', 'purple')

# Add a title for the entire figure
fig.suptitle('Screen Time Distribution by Activity and Day', fontsize=16, fontweight='bold')

plt.tight_layout()
plt.show()
```

## Hypothesis 1: Investigate if there is a significant difference in well-being scores between individuals with high and low screen time.

```python
# Calculate total and average screen time
merged_data['Total_Screen_Time'] = merged_data[['C_wk', 'C_we', 'G_wk', 'G_we', 'S_wk', 'S_we', 'T_wk', 'T_we']].sum(axis=1)
merged_data['Avg_Screen_Time'] = merged_data['Total_Screen_Time'] / 7


# Categorisse into high and low screen time
median_screen_time = merged_data['Avg_Screen_Time'].median()
merged_data['Screen_Time_Group'] = np.where(merged_data['Avg_Screen_Time'] > median_screen_time, 'High', 'Low')
```

```python
# Categorise into high and low screen time
median_screen_time = merged_data['Avg_Screen_Time'].median()
merged_data['Screen_Time_Group'] = np.where(merged_data['Avg_Screen_Time'] > median_screen_time, 'High', 'Low')

# Define high and low screen time groups
high_group = merged_data[merged_data['Screen_Time_Group'] == 'High']['Avg_Screen_Time']
low_group = merged_data[merged_data['Screen_Time_Group'] == 'Low']['Avg_Screen_Time']

import scipy.stats as stats

# Perform the Mann-Whitney U Test (non-parametric test)
u_stat, p_value = stats.mannwhitneyu(high_group, low_group)

# Interpret results
print(f"P-value: {p_value}")
if p_value < 0.05:
    print("There is a significant difference in well-being scores between high and low screen time groups.")
else:
    print("There is no significant difference in well-being scores between high and low screen time groups.")
```

```python

import scipy.stats as st

# Merge datasets
merged_data = pd.merge(dataset1, dataset2, on='ID')
merged_data = pd.merge(merged_data, dataset3, on='ID')

# WELLBEING DATA FRAME
wellbeing = merged_data[['Optm', 'Usef', 'Relx', 'Intp', 'Engs', 'Dealpr', 'Thcklr', 'Goodme', 'Clsep', 'Conf', 'Mkmind', 'Loved', 'Intthg', 'Cheer']].values

# SMARTPHONE WEEKEND DATA FRAME
s_we = merged_data['S_we'].values
s_we_median = np.median(s_we)  # Fixed: use np.median instead of stats.median
s_we_high = s_we[s_we > s_we_median]
s_we_high_wellbeing = wellbeing[s_we > s_we_median]

# Compute smartphone group statistics
s_we_high_wellbeing_means = np.mean(s_we_high_wellbeing, axis=0)
s_we_high_wellbeing_avg = np.mean(s_we_high_wellbeing_means)
s1 = np.std(s_we_high, ddof=1)
n1 = len(s_we_high)
var1 = np.var(s_we_high, ddof=1)

print(f"Sample size (high smartphone use): {n1}")
print(f"Mean well-being score (smartphone group): {s_we_high_wellbeing_avg:.2f}")
print(f"Standard deviation (smartphone group): {s1:.2f}")
print(f"Variance (smartphone group): {var1:.2f}")

# TV WEEKEND DATA FRAME
t_we = merged_data['T_we'].values
t_we_median = np.median(t_we)
t_we_high = t_we[t_we > t_we_median]
t_we_high_wellbeing = wellbeing[t_we > t_we_median]

# Compute TV group statistics
t_we_high_wellbeing_means = np.mean(t_we_high_wellbeing, axis=0)
t_we_high_wellbeing_avg = np.mean(t_we_high_wellbeing_means)
s2 = np.std(t_we_high, ddof=1)
n2 = len(t_we_high)
var2 = np.var(t_we_high, ddof=1)

print(f"Sample size (high TV use): {n2}")
print(f"Mean well-being score (TV group): {t_we_high_wellbeing_avg:.2f}")
print(f"Standard deviation (TV group): {s2:.2f}")
print(f"Variance (TV group): {var2:.2f}")

# Perform t-test
t_stats, p_val = st.ttest_ind_from_stats(s_we_high_wellbeing_avg, s1, n1, t_we_high_wellbeing_avg, s2, n2, equal_var=False)

print(f"T-statistic: {t_stats:.2f}")
print(f"P-value: {p_val:.5f}")

if p_val < 0.05:
    print("There is a statistically significant difference in well-being scores between high smartphone and TV use on weekends.")
else:
    print("There is no statistically significant difference in well-being scores between high smartphone and TV use on weekends.")
```

```python
import statistics as stats

# WELLBEING DATA FRAME
wellbeing = merged_data[['Optm', 'Usef', 'Relx', 'Intp', 'Engs', 'Dealpr', 'Thcklr', 'Goodme', 'Clsep', 'Conf', 'Mkmind', 'Loved', 'Intthg', 'Cheer']].values

# SMARTPHONE WEEKENDS DATA FRAME
s_we = merged_data['S_we'].values                                                                           # IMPORT VALUES
s_we_mean = st.tmean(s_we)                                                                                  # SMARTPHONE WEEKENDS SAMPLE MEAN
s_we_median = stats.median(s_we)                                                                            # SMARTPHONE WEEKENDS SAMPLE MEDIAN

print("Descriptive Statistics for Smartphone Usage on Weekends (Group 1):")


s_we_high = s_we[s_we > s_we_median]                                                                        # SEPARATE HIGH SMARTPHONE ON WEEKENDS USERS - GROUP 1
s_we_high_mean = st.tmean(s_we_high)
s_we_high_median = stats.median(s_we_high)
s_we_high_wellbeing = wellbeing[s_we > s_we_median]                                                         # SEPARATE WELLBEING GROUP 1
s_we_high_wellbeing_means = st.tmean(s_we_high_wellbeing, axis=0)                                           # COMPUTE MEANS FOR EACH INDIVIDUAL WELLBEING SCORE GROUP 1
s_we_high_wellbeing_avg = st.tmean(s_we_high_wellbeing_means)                                               # COMPUTE A SINGLE MEAN SCORE FOR ALL WELLEBING SCORES GROUP 1
s1 = st.tstd(s_we_high, ddof=1)                                                                             # STANDARD DEVIATION FOR GROUP 1
n1 = len(s_we_high)                                                                                         # LENGTH OF SAMPLE FOR GROUP 1
var1 = np.var(s_we_high, ddof=1)                                                                            # VARIANCE FOR GROUP 1
ddof1 = n1 - 1                                                                                              # DEGREES OF FREEDOM FOR GROUP 1  

print("Mean smartphone use on weekends for the high use subset: %.2f hours." % s_we_high_mean)
print("Median smartphone use for Group 1: %.2f hours." % s_we_high_median)
print("Sample size for Group 1: %d." % n1)
print("Degrees of Freedom for Group 1: %d." % ddof1)
print("Standard Deviation for Group 1: %.2f."% s1)
print("Variance for Group 1: %.2f." % var1)
print("Mean score for the collective wellbeing scores for Group 1: %.2f." % s_we_high_wellbeing_avg)

# TV WEEKENDS DATA FRAME
t_we = merged_data['T_we'].values                                                                           # IMPORT VALUES
t_we_mean = st.tmean(t_we)                                                                                  # TV WEEKENDS SAMPLE MEAN
t_we_median = stats.median(t_we)                                                                            # TV WEEKENDS SAMPLE MEDIAN

print("Descriptive Statistics for TV Usage on Weekends (Group 2):")

t_we_high = t_we[t_we > t_we_median]                                                                        # SEPARATE HIGH TV ON WEEKENDS USERS - GROUP 2
t_we_high_mean = st.tmean(t_we_high)
t_we_high_median = stats.median(t_we_high)
t_we_high_wellbeing = wellbeing[t_we > t_we_median]                                                         # SEPARATE WELLBEING GROUP 2
t_we_high_wellbeing_means = st.tmean(t_we_high_wellbeing, axis=0)                                           # COMPUTE MEANS FOR EACH INDIVIDUAL WELLBEING SCORE GROUP 2
t_we_high_wellbeing_avg = st.tmean(t_we_high_wellbeing_means)                                               # COMPUTE A SINGLE MEAN SCORE FOR ALL WELLEBING SCORES GROUP 2
s2 = st.tstd(t_we_high, ddof=1)                                                                             # STANDARD DEVIATION FOR GROUP 2
n2 = len(t_we_high)                                                                                         # LENGTH OF SAMPLE FOR GROUP 2
var2 = np.var(t_we_high, ddof=1)                                                                            # VARIANCE FOR GROUP 2
ddof2 = n2 - 1                                                                                              # DEGREES OF FREEDOM FOR GROUP 2   

print("Mean TV use on weekends for the high use subset: %.2f hours." % t_we_high_mean)
print("Median TV use for high use subset: %.2f hours." % t_we_high_median)
print("Sample size for Group 2: %d." % n2)
print("Degrees of Freedom: %d." % ddof2)
print("Standard Deviation for Group 2: %.2f." % s2)
print("Variance for Group 2: %.2f." % var2)
print("Mean score for collective wellbeing scores for Group 2: %.2f." % t_we_high_wellbeing_avg)

# T-test
t_stats, p_val = st.ttest_ind_from_stats(s_we_high_wellbeing_avg, s1, n1, t_we_high_wellbeing_avg, s2, n2, equal_var=False)
print("Two Sample t-Test Results:")
print("T-statistic: ." , t_stats)
print("P-value: .", p_val)
print('\n')

print("Outcome:")
if p_val < 0.05:
        print("There is a statistically significant difference in well-being scores between high smartphone and TV use on weekends.")
else:
        print("There is no statistically significant difference in well-being scores between high smartphone and TV use on weekends.")
```
